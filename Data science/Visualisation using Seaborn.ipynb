#### Scatter plot
The scatter plot is a mainstay of statistical visualization that shows joint distribution of two variables using a cloud of points, where each point represents an observation in the dataset
We can use the `scatterplot()` or `replot()` to plot scatter plot
`scatterplot()` is the default function

```python
sns.scatterplot(x='cholera',y='dysentery',data=dataset)
```

```python
# adding some additional parameters
#adding the breakdown
sns.scatterplot(x='cholera',y='dysentery',hue='province', data=dataset)
```

```python
# adding some additional parameters
#adding the breakdown and the size of the circle
sns.scatterplot(x='cholera',y='dysentery',hue='province', size='measles', data=dataset)
```

As I said above the `scatterplot()` is the default kind in `relplot()` (it can also be forced by setting `kind="scatter"`):

See difference

```python
sns.relplot(x='cholera',y='dysentery',hue='province', size='measles', kind="scatter", data=dataset)
```

### Plotting with Categorical data

In seaborn, there are several different ways to visualize a relationship involving categorical data. Similar to the relationship between `relplot()` and either `scatterplot()` or `lineplot()`, there are two ways to make these plots. There are a number of axes-level functions for plotting categorical data in different ways and a figure-level interface, `catplot()`, that gives unified higher-level access to them.

#### Categorical scatterplots:
stripplot() (with kind=strip; the default)

swarmplot() (with kind=swarm)

#### Categorical distribution plots:
boxplot() (with kind=box)

violinplot() (with kind=violin)

boxenplot() (with kind=boxen)

#### Categorical estimate plots:
pointplot() (with kind=point)

barplot() (with kind=bar)

countplot() (with kind=count)



```python
#indexing data
data_f= dataset[0:40]
sns.catplot(x="dysentery",y="province",data=data_f)
```

```python
#Let us now say we want o show swarm plot 
#This plot avoids overlaping
sns.catplot(x="dysentery",y="province",kind="swarm",data=data_f)
```

```python
#Box plot
sns.catplot(x="dysentery",y="province",kind="box",data=data_f)
```

When adding a hue semantic, the box for each level of the semantic variable is moved along the categorical axis so they dont overlap:

```python
data=dataset[0:20]
sns.catplot(x="dysentery",y="province",kind="box",hue="district",data=data)
```

### Violin plots

It is similar to a box plot, with the addition of a rotated kernel density plot on each side. 

```python
sns.catplot(x="dysentery",y="province",kind="violin",data=data_f)
```

### Bar plots


```python
sns.catplot(x="province",y="dysentery",kind="bar",data=data_f)
#You can also use barplot() to plot the same graph
#sns.barplot(x="province",y="dysentery",data=data_f)
```
